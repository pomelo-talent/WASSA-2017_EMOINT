```python
import pandas as pd
anger_ml_score = pd.read_csv("score_anger.csv")
anger_ml_score.index = ['anger','anger']
fear_ml_score = pd.read_csv("score_fear.csv")
fear_ml_score.index = ['fear','fear']
joy_ml_score = pd.read_csv("score_joy.csv")
joy_ml_score.index = ['joy','joy']
sadness_ml_score = pd.read_csv("score_sadness.csv")
sadness_ml_score.index = ['sadness','sadness']
whole_ml_score = pd.read_csv("score_whole.csv")
whole_ml_score.index = ['whole','whole']

anger_dl_score = pd.read_csv("score_lstmatt_anger.csv")
anger_dl_score.index = ['anger']
fear_dl_score = pd.read_csv("score_lstmatt_fear.csv")
fear_dl_score.index = ['fear']
joy_dl_score = pd.read_csv("score_lstmatt_joy.csv")
joy_dl_score.index = ['joy']
sadness_dl_score = pd.read_csv("score_lstmatt_sadness.csv")
sadness_dl_score.index = ['sadness']
whole_dl_score = pd.read_csv("score_lstmatt_whole.csv")
whole_dl_score.index = ['whole']
```

```python
score =  pd.concat([anger_ml_score, fear_ml_score, joy_ml_score, sadness_ml_score, whole_ml_score,\
                    anger_dl_score, fear_dl_score, joy_dl_score, sadness_dl_score, whole_dl_score], axis=0)
score_f = score.rename(columns={'Unnamed: 0':'Model'})
score_f.sort_index(inplace=True)
score_f
```
